### Way to make data easier to read
One thing that would probably make our data a little easier to work with would be to make the "Date" row our column index and we could call each column name something like "yyyy_mm" to make it crystal clear that it is the average price in the borough for a given year and month

This way, when we melt the data and transform it to long format, it will be nicely labeled for us as well

```python
#import datetime module from datetime
from datetime import datetime

#Create a list that will replace our column titles with the proper format
headerList =[]

for colCounter in range(1,len(properties_cleanedtrans.columns)):
    #Extract the "Date" and convert to string
    varDateString = str(properties_cleanedtrans.iloc[0,colCounter])

    #Extract relevant portion of string
    yearMonth = varDateString[0:10]
    
    #Append to headerList
    headerList.append(yearMonth)
    
#Create a simple list to make a zip into a dictionary
keyList = list(range(1,len(properties_cleanedtrans.columns)+1))
#print(keyList)

#Create new dictionary for replacements
newHeaders = dict(zip(keyList,headerList))

#Change Column Name & Remove Date Row
properties_cleanedtrans.rename(columns=newHeaders,inplace = True)
```

```python
#Finally Remove the Date row of observatiosn since it is now redundant
properties_cleanedtrans.drop('Date', axis=0, inplace= True)
properties_cleanedtrans
```

#### Collapsing Data
*Based on what this question is asking and I'm trying to infer from the language, it would appear we want to collapse our data around years using the average (so for 2000 for a given borough, we'd take the average price of 12 points (JAN thru DEC for 2000) and that would become the average price for 2000*

From this it would appear it would make the most sense to make a new dataframe with the contained values for this.

```python
meanList = [1,2,3,4,5]
print(np.mean(meanList))
```

```python
#First, let's make a new column that has just the year
properties_cleanedtrans_long["Year"] = pd.DatetimeIndex(properties_cleanedtrans_long["Date"]).year

#Second, let's make a list of the boroughs
boroughtList = list(properties_cleanedtrans_long["Borough"].unique())

#Create a new data frame that will hold our target price calculations
properties_year = pd.DataFrame(columns = ["Borough" , "avgPrice1998", "avgPrice2018"])
properties_year["Borough"] = boroughtList

#Initialize an empty list that will hold our averages and create a list of averages for 2018
avg2018Holder = []
for idx, boroughName in enumerate(boroughtList):
    borough2018PriveAvgList = properties_cleanedtrans_long.loc[(properties_cleanedtrans_long["Borough"] == boroughName) & (properties_cleanedtrans_long["Year"] == 2018),:]
    borough2018PriveAvg = np.mean(list(borough2018PriveAvgList["AvgPrice"]))
    avg2018Holder.append(borough2018PriveAvg)

properties_year["avgPrice2018"] = avg2018Holder

#Initialize an empty list that will hold our averages and create a list of averages for 1998
avg1998Holder = []
for idx, boroughName in enumerate(boroughtList):
    borough1998PriveAvgList = properties_cleanedtrans_long.loc[(properties_cleanedtrans_long["Borough"] == boroughName) & (properties_cleanedtrans_long["Year"] == 1998),:]
    borough1998PriveAvg = np.mean(list(borough1998PriveAvgList["AvgPrice"]))
    avg1998Holder.append(borough1998PriveAvg)

properties_year["avgPrice1998"] = avg1998Holder

properties_year
```

*I went through effort to make a nice table to do the modeling (or more specifically to measure the value of change since we now have a nice dataframe that is condensed and cleaned). In addition, it makes it a lot easier to type things out*
